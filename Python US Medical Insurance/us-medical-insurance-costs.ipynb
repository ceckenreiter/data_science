```python
#import csv library
import csv
```

The first step of analyzing any data is to see what we are working with. For this I can use csv methods such as DictReader to see what data is available to me. 

From a quick peak at the data I can see we have 7 categories:
    1.) Age - string 
    2.) Sex - string
    3.) bmi - string
    4.) children - string
    5.) smoker - string
    6.) region - string
    7.) charges - string 

I also know that there is no missing data from this table. I will organize each column into its own lists which will make it easier to utilize and analyze. For columns that hold numerical data as strings, I will convert them to float values to make future computations easier 

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []


with open('insurance.csv', newline = '') as insurance_info: 
    var2 = csv.DictReader(insurance_info)
    for row in var2:
        age.append(float(row['age']))
        sex.append(row['sex'])
        bmi.append(float(row['bmi']))
        children.append(float(row['children']))
        smoker.append(row['smoker'])
        region.append(row['region'])
        charges.append(float(row['charges']))
```

```python
csv_length = len(age)

def average(list):
    length = len(list)
    total = 0
    for number in list:
        total += number
    average = total/length 
    print(average)


def find_options(list):
    variables = []
    for item in list: 
        if variables.count(item) == 0:
            variables.append(item)       
    return variables
    

def count(list):
    variables = find_options(list)
    statements = []
    for var in variables:
        total = []
        for item in list:
            if item == var:
                total.append(item)
        statements.append([var, len(total)])
    for pair in statements:
        print(str(pair[1]) + ' of ' + str(pair[0]))


sex_vs_charges =  list(zip(sex, charges))
region_vs_charges = list(zip(region, charges))
sex_vs_smoker = list(zip(sex, smoker))
age_vs_smoker = list(zip(age, smoker))
```
