```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

starting_df = pd.read_csv('all_data.csv')
starting_df['GDP'] = starting_df.GDP.apply(lambda x: x/1e12)
starting_df.rename(columns = {"Life expectancy at birth (years)": "Life Expectancy"}, inplace = True)

Chile = starting_df[starting_df.Country == 'Chile']
Germany = starting_df[starting_df.Country == 'Germany']
China = starting_df[starting_df.Country == 'China']
Mexico = starting_df[starting_df.Country == 'Mexico']
US = starting_df[starting_df.Country == 'United States of America']
Zimbabwe = starting_df[starting_df.Country == 'Zimbabwe']
```

```python
Chile_GDP_Year = [Chile.GDP.tolist(),Chile.Year.tolist()]
Germany_GDP_Year = [Germany.GDP.tolist(), Germany.Year.tolist()]
China_GDP_Year = [China.GDP.tolist(), China.Year.tolist()]
Mexico_GDP_Year = [Mexico.GDP.tolist(), Mexico.Year.tolist()]
US_GDP_Year = [US.GDP.tolist(), US.Year.tolist()]
Zimbabwe_GDP_Year = [Zimbabwe.GDP.tolist(), Zimbabwe.Year.tolist()]

plt.plot(Chile_GDP_Year[1], Chile_GDP_Year[0])
plt.plot(Germany_GDP_Year[1], Germany_GDP_Year[0])
plt.plot(China_GDP_Year[1], China_GDP_Year[0])
plt.plot(Mexico_GDP_Year[1], Mexico_GDP_Year[0])
plt.plot(US_GDP_Year[1], US_GDP_Year[0])
plt.plot(Zimbabwe_GDP_Year[1], Zimbabwe_GDP_Year[0])
plt.xlabel('Year')
plt.ylabel('GDP (in trillions of USD)')
plt.xticks([2000, 2002,2004,2006,2008,2010,2012,2014], rotation=45)
plt.title('GDP of Various Countries from 2000 - 2015')
plt.legend(['Chille','Germany','China','Mexico', 'US', 'Zimbabwe'], loc=(1.05, .4))
plt.show()
plt.close()

```

```python

countries_array = starting_df.Country.unique()
index=1
color=['', 'red', 'green', 'blue', 'orange', 'brown', 'pink']


for item in countries_array: 
    life_year = starting_df[starting_df.Country == item]
    plt.subplot(2,3,index)
    plt.plot(life_year.Year, life_year['Life Expectancy'], color=color[index])
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy at Birth (Years)')
    plt.title(item + ' Life Expectancy 2000 - 2015')
    plt.xticks([2000, 2005, 2010, 2015])
    index=index+1
plt.subplots_adjust(right=2,top=2)
plt.show()
plt.clf()

for item in countries_array: 
    life_year = starting_df[starting_df.Country == item]
    plt.scatter(life_year.Year, life_year['Life Expectancy'])
plt.legend(countries_array, loc=(1.04, .5))
plt.title('Life Expectancy of Various Countries from 2000-2015')
plt.ylabel('Life Expectancy at Birth (Years)')
plt.xlabel('Years')
plt.show()
plt.clf()

                  
```
